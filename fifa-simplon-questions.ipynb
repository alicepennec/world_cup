```python
# 1. quel pays a gagné le + de matchs en coupe du monde ?
combined_df['winner'] = combined_df.apply(
    lambda x: x['home_team'] if x['home_result'] > x['away_result'] else 
              (x['away_team'] if x['away_result'] > x['home_result'] else 'Draw'), 
    axis=1
)
win_count = combined_df[combined_df['winner'] != 'Draw'].groupby('winner').size().reset_index(name='win_count').sort_values(by='win_count', ascending=False)
most_wins_team = win_count.head()
print("quel pays a gagné le + de matchs en coupe du monde ?\n",most_wins_team)

# 2. quel est le top 5 des vainqueurs de coupe du monde(finale) ?
combined_df.loc[(combined_df['édition'] == '1994-USA') & (combined_df['round'] == 'Final'), 'winner'] = 'Brazil'
combined_df.loc[(combined_df['édition'] == '2006-GERMANY') & (combined_df['round'] == 'Final'), 'winner'] = 'Italy'
combined_df.loc[(combined_df['édition'] == '2022-QATAR') & (combined_df['round'] == 'Final'), 'winner'] = 'Argentina'

tmp = combined_df[combined_df['round'] == 'Final']
win_count = tmp[tmp['winner'] != 'Draw'].groupby('winner').size().reset_index(name='win_count').sort_values(by='win_count', ascending=False)
most_wins_team = win_count.head()
print("quel est le top 5 des vainqueurs de coupe du monde(finale) ?\n",most_wins_team)

# 3. quel est le plus grand écart de buts dans un match ?
combined_df['goal_diff'] = abs(combined_df['home_result'] - combined_df['away_result'])
max_goal_diff_match = combined_df.loc[combined_df['goal_diff'].idxmax()]
print("quel est le plus grand écart de buts dans un match ? \n",max_goal_diff_match)

# 4. combien de matchs nuls au total ?
total_draws = combined_df[combined_df['winner'] == 'Draw'].shape[0]
print("combien de matchs nuls au total ? \n",total_draws)

# 5. quel pays est allé le plus souvent en phase finale de cdm ?
final_teams = pd.concat([combined_df[combined_df['round'] == 'Final']['home_team'], combined_df[combined_df['round'] == 'Final']['away_team']])
most_final_appearances = final_teams.value_counts().idxmax()
print("quel pays est allé le plus souvent en phase finale de cdm ? \n",most_final_appearances)
```
